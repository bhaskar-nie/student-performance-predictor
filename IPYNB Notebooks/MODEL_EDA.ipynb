### Problem statement
This project aims to understand how a student's performance (Percentage in Test 3) is affected by other variables such as Gender, Learning Support, Parent's Education, Had Lunch Before Exams, Course Completed, and marks in Percentage in Test 1 and Percentage in Test 2.

### Dataset information

- **Gender**: The sex of the students (Male/Female)  
- **Learning Support**: The type of learning support students receive (Limited Tutoring, Occasional Mentoring, Standard School Support, Private Coaching, Elite Academic Support)

- **Course Completed**: Indicates whether the student completed the test preparation course before taking the test (Complete or None)  
- **Percentage in Test 1**: The score achieved in the first test 
- **Percentage in Test 2**: The score achieved in the second test
- **Percentage in Test 3**: The score achieved in the third test

### Adding columns for "Total Marks" and "Average Marks

```python
# Adding columns for Total Marks and Average Marks
df['Total Marks'] = df['Percentage in Test 1'] + df['Percentage in Test 2'] + df['Percentage in Test 3']
df['Average Marks'] = df['Total Marks'] / 3

# Displaying the first few rows to verify the changes
df.head()
```

```python
full_marks_in_test_1 = df[df['Percentage in Test 1'] == 100]['Average Marks'].count()
full_marks_in_test_2 = df[df['Percentage in Test 2'] == 100]['Average Marks'].count()
full_marks_in_test_3 = df[df['Percentage in Test 3'] == 100]['Average Marks'].count()

print(f'Number of students with full marks in Test 1: {full_marks_in_test_1}')
print(f'Number of students with full marks in Test 2: {full_marks_in_test_2}')
print(f'Number of students with full marks in Test 3: {full_marks_in_test_3}')

```

```python
less_than_40_test1 = df[df['Percentage in Test 1'] <= 40]['Average Marks'].count()
less_than_40_test2 = df[df['Percentage in Test 2'] <= 40]['Average Marks'].count()
less_than_40_test3 = df[df['Percentage in Test 3'] <= 40]['Average Marks'].count()

print(f'Number of students with less than 40 marks in Test 1: {less_than_40_test1}')
print(f'Number of students with less than 40 marks in Test 2: {less_than_40_test2}')
print(f'Number of students with less than 40 marks in Test 3: {less_than_40_test3}')
```

#### 4.3 Multivariate analysis using pieplot

```python
plt.rcParams['figure.figsize'] = (30, 12)

plt.subplot(1, 5, 1)
size = df['Gender'].value_counts()
labels = df['Gender'].unique()
color = ['red', 'green']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')
plt.title('Gender', fontsize=20)
plt.axis('off')

plt.subplot(1, 5, 2)
size = df['Learning Support'].value_counts()
labels = df['Learning Support'].unique()
color = ['red', 'green', 'blue', 'cyan', 'orange']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')
plt.title('Learning Support', fontsize=20)
plt.axis('off')

plt.subplot(1, 5, 3)
size = df['Had Lunch'].value_counts()
labels = df['Had Lunch'].unique()
color = ['red', 'green']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')
plt.title('Had Lunch', fontsize=20)
plt.axis('off')

plt.subplot(1, 5, 4)
size = df['Course Completed'].value_counts()
labels = df['Course Completed'].unique()
color = ['red', 'green']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')
plt.title('Course Completed', fontsize=20)
plt.axis('off')

plt.subplot(1, 5, 5)
size = df['Parent\'s Education'].value_counts()
labels = df['Parent\'s Education'].unique()
color = ['red', 'green', 'blue', 'cyan', 'orange', 'grey']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')
plt.title('Parent\'s Education', fontsize=20)
plt.axis('off')

plt.tight_layout()
plt.grid()

plt.show()
```

#####  Insights
- The number of Male and Female students is almost equal.
- The greatest number of students belong to **Occasional Mentoring**.
- The number of students who have lunch before exams is greater.
- A greater number of students have **NOT Completed the Course**.
- The largest group of students' parents have attained **Bachelor's Level** education, followed closely by **Undergraduate**.

#### BIVARIATE ANALYSIS ( Does Learning Support have any impact on student's performance ? )

```python
# Group data by 'Learning Support'
Group_data2 = df.groupby('Learning Support')

# Create subplots
f, ax = plt.subplots(1, 3, figsize=(20, 8))

# Plot for Test 1
sns.barplot(x=Group_data2['Percentage in Test 1'].mean().index, 
            y=Group_data2['Percentage in Test 1'].mean().values, 
            palette='flare', ax=ax[0])
ax[0].set_title('Percentage in Test 1', color='#005ce6', size=20)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=15)

# Plot for Test 2
sns.barplot(x=Group_data2['Percentage in Test 2'].mean().index, 
            y=Group_data2['Percentage in Test 2'].mean().values, 
            palette='coolwarm', ax=ax[1])
ax[1].set_title('Percentage in Test 2', color='#005ce6', size=20)
for container in ax[1].containers:
    ax[1].bar_label(container, color='black', size=15)


# Plot for Test 3
sns.barplot(x=Group_data2['Percentage in Test 3'].mean().index, 
            y=Group_data2['Percentage in Test 3'].mean().values, 
            palette='mako', ax=ax[2])
ax[2].set_title('Percentage in Test 3', color='#005ce6', size=20)
for container in ax[2].containers:
    ax[2].bar_label(container, color='black', size=15)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

#### Insights 
- Elite Academic Support students have scored the highest marks.
- Limited Tutoring  students have scored the lowest marks.
